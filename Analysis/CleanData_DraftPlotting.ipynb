```python
# Bar Charts
crime_outcomes = ["seriousviolent", "injury", "treatment"]
weight = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(weight))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis, weight, color="b", align="center")        
```

```python
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, crime_outcomes)
```

```python
# Set the limits of the x axis
plt.xlim(-0.75, len(x_axis)-0.25)
```

```python
# Set the limits of the y axis
plt.ylim(0, max(weight)+0.4)
```

```python
# Give the chart a title, x label, and y label
plt.title("Weight (Estimate of Victimization) vs Crime outcome")
plt.xlabel("Crime_outcome")
plt.ylabel("Weight (Estimate of Victimization)")
plt.show
```
